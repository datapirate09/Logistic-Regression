```python
import pandas as pd
import matplotlib.pyplot as plt
hearing_test_dataset = pd.read_csv("Your Dataset Path",low_memory=False)
print(hearing_test_dataset.head())
```

```python
x1_overall_features = hearing_test_dataset["age"].to_numpy()
x2_overall_features = hearing_test_dataset["physical_score"].to_numpy()
y_overall_features = hearing_test_dataset["test_result"].to_numpy()
plt.plot(x1_overall_features, y_overall_features, 'o')
plt.xlabel('Age')
plt.ylabel('Test Result')
plt.show()
plt.plot(x2_overall_features, y_overall_features, 'x')
plt.xlabel('Physical Score')
plt.ylabel('Test Result')
plt.show()
```

```python
input_features = ["age", "physical_score"]
output_features = ["test_result"]
X = hearing_test_dataset[input_features]
Y = hearing_test_dataset[output_features]
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_Y = StandardScaler()
y_train = scaler_Y.fit_transform(y_train) 
y_test = scaler_Y.transform(y_test)
```

```python
import numpy as np
no_of_observations, no_of_features = X_train.shape
weights_global = np.random.randn(no_of_features, 1) * 0.01
bias_global = 0
```

```python
def sigmoid(sigmoid_input):
    return 1 / (1 + np.exp(-sigmoid_input))
```

```python
def compute_cost(X,y,w,b):
    #this function should compute the log loss function
    cost = 0.0
    for i in range(no_of_observations):
        z_i = np.dot(X[i],w) + b
        f_wb_i = sigmoid(z_i)
        cost += (-y[i]*np.log(f_wb_i)-(1-y[i])*np.log(1-f_wb_i))
    cost = cost/no_of_observations
    return cost
```

```python
from sklearn.metrics import accuracy_score

# Convert y_test to binary (if it's continuous)
y_test_binary = (y_test >= 0.5).astype(int)  # Convert to 0 or 1

# Compute predictions using trained weights and bias
z_test = np.dot(X_test, weights_global) + bias_global  # Linear combination
y_pred_prob = 1 / (1 + np.exp(-z_test))  # Apply sigmoid function

# Convert probabilities into binary predictions (0 or 1)
y_pred_binary = (y_pred_prob >= 0.5).astype(int)  

# Compute accuracy
accuracy = accuracy_score(y_test_binary, y_pred_binary)
print(f"Accuracy of the model: {accuracy:.4f}")
```
